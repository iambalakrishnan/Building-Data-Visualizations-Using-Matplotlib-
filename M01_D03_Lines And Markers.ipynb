```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
x = np.linspace(start=0, stop=10, num=50)
```

```python
plt.plot(x, np.sin(x), label = 'sine curve')
plt.plot(x, np.cos(x), label = 'cosine curve')
plt.legend()
plt.title('Sine and Cosine curve with respected x value')
plt.show()
```

```python
plt.plot(x, np.sin(x), label = 'sine curve')
plt.plot(x, np.cos(x), label = 'cosine curve')

plt.legend()
plt.title('Sine and Cosine curve with respected x value')
plt.show()
```

```python
# We can also customize color 

plt.plot(x, np.sin(x), label = 'sine curve', color='green')#green color by full name
plt.plot(x, np.cos(x), label = 'cosine curve', color='m')#magenda color by first letter of the color

plt.legend()
plt.title('Sine and Cosine curve with respected x value')
plt.show()
```

```python
random_array = np.random.randn(20)
```

```python
random_array
```

```python
# x coordinate will be index of the y coordinate
plt.plot(random_array, color='m')#by default it assign as y co ordinate
plt.show()
```

```python
# we can also customize lise style 
#linestyle = : --for dotted line and the default line is solid line 

plt.plot(random_array, color='b', linestyle = ':')
plt.show()
```

```python
# We can also draw dash line

plt.plot(random_array, color='m', linestyle = '--')
plt.show()
```

```python
#we can also customize thickness of the line, the default linewidth is set to 1

plt.plot(random_array, color='black', linestyle = '--', linewidth = 3)
plt.show()
```

```python
#Use dimond shaped markers to donate  the points in our plot
#The color attribute set both line and marker color
plt.plot(random_array, color='black', marker = 'd')
plt.show()
```

```python
# We can make our marker more prominent by specifing marker size object 
#The default size of the marker is 6
plt.plot(random_array, color='black', marker = 'd', markersize = 10)
plt.show()
```

```python
# If you only want to specify marker not to specify line you can set linestyle=none
plt.plot(random_array, color='black', marker = 'd', markersize = 10, linestyle = 'None')
plt.show()
```

<h4> Scatter plot
    

Scatter plot are used to model relationship between two variables, one on the X axis and one on the Y axis 

```python
plt.scatter(range(0,20), random_array, color='m', marker='d')
plt.show()
```
