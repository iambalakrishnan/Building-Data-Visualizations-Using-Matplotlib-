<h3>Figures and Axes
    


A figure in matplotlib is a high-level container which represent the window or the drawing area where your plots are drawn

```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
#You can instantiate a figure by calling pyplot.figure
#you can draw axes within the figure 

#The add_axes function  on the figure add one set of x and y axes to the figure
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
#The input argument inside the add_axes denotes how exactly axes will be drawn with in the larger figure container 

plt.show()
```

```python
type(ax)
```

```python
# You can draw multiple axes with in the same figure 

fig = plt.figure()

ax = fig.add_axes([0, 0.6, 1, 0.4])
ax2 = fig.add_axes([0, 0, 0.8, 0.4])

plt.show()
```

```python
x = np.linspace(start=0, stop=10, num=50)
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0.2, 0.6, 1, 0.4])
ax2 = fig.add_axes([0, 0, 0.8, 0.4])

ax1.plot(x, np.sin(x))
ax1.set_xlabel('x', fontsize=15, color='r')
ax1.set_ylabel('sin(x)', fontsize=15, color='r')

ax2.plot(x, np.cos(x))
ax2.set_xlabel('x', fontsize=15, color='r')
ax2.set_ylabel('cos(x)', fontsize=15, color='r')


plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.5, 0.5, 0.4, 0.4])
plt.show()
```

```python
# We can also se t fig size

fig = plt.figure(figsize=(8,8))

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.5, 0.5, 0.4, 0.4])
plt.show()
```

```python
#Subplot

fig = plt.figure(figsize=(8,8))

ax1 = fig.add_subplot(221)

ax1.plot([2,4,6,8],[4,8,12,16])

ax2=fig.add_subplot(222)
ax2.plot(x, np.sin(x))

ax2=fig.add_subplot(224)
ax2.plot(x, np.cos(x))

plt.show()
```

```python
ax1 = plt.subplot2grid((2,3),(0,0))
ax1.plot(x, np.sin(x))
ax1.set_label('sine curve')
ax1.set_title('sine curve')

ax2 = plt.subplot2grid((2,3),(0,1))
ax2.plot(x, np.cos(x))
ax2.set_label('cosine curve')
ax2.set_title('cos curve')

ax3 = plt.subplot2grid((2,3),(0,2), rowspan=2)
ax3.plot([2,4,6,8],[4,8,12,16])
ax3.set_label('Straight Line')
ax3.yaxis.tick_right()
ax3.set_title('Straight Line')

ax4 = plt.subplot2grid((2,3),(1,0), colspan=2)
ax4.plot(x, np.exp2(x))
ax4.set_label('Exponential curve')
ax4.set_title('Exponential curve')
```
