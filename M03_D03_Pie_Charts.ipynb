```python
import numpy as np
import pandas as pd
import matplotlib 
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('sector_weighting.csv')
data.head()
```

```python
data.shape
```

Assume this contain the breakdown of the investment portfolio of a fictitious mutual fund

```python
plt.pie(data['Percentage'],labels=data['Sector'])

plt.show()
```

```python
plt.pie(data['Percentage'],labels=data['Sector'])

plt.axis('equal')#Display the pie chart in perfect circular format
plt.show()
```

```python
#Lets customize the colors in the pie chart 
#Create a list of colors

colors = ['deeppink','aqua','magenta','silver','lime']

plt.pie(data['Percentage'],labels=data['Sector'], colors = colors, autopct = '%.2f')
#Colors set to the colors for each sector autopct represent the format for the displayed value

plt.axis('equal')#Display the pie chart in perfect circular format
plt.show()
```

```python
colors = ['deeppink','aqua','magenta','silver','lime']

plt.pie(data['Percentage'],labels=data['Sector'], colors = colors, autopct = '%.2f', 
        startangle=90, counterclock=False)
#The pie widges start the first widge at 90 degree
#counterclock=False set to false and the widges will be layed out clock wise

plt.axis('equal')#Display the pie chart in perfect circular format
plt.show()
```

```python
#Explode allows  us to highlight a wedge in the pie chart by saperating it from the pie by a fraction of its radius

explode=(0, 0.1, 0, 0.3, 0)

plt.pie(data['Percentage'],labels=data['Sector'], colors = colors, autopct = '%.2f', explode=explode)
plt.show()
```

```python
#View the component of the pie chart

wedges, texts, autotexts = plt.pie(data['Percentage'],colors = colors, autopct = '%.2f', explode=explode)

plt.axis('equal')

print('Wedges : ',wedges)#List of patch object each patch object represent one sector in our pie chart
print('Texts : ', texts)#all the text labels along with their positions in the pie chart
print('Autotexts : ', autotexts)#Represent the numerica values only if autopct is set for the pie chart
```

```python
#Format some component of the pie chart


wedges, texts, autotexts = plt.pie(data['Percentage'],labels=data['Sector'],colors = colors, autopct = '%.2f', explode=explode)

plt.axis('equal')

plt.legend(title='Sector', loc='upper center')

wedges[1].set(edgecolor='blue', linewidth=3)
wedges[2].set(edgecolor='red', linewidth=3)
wedges[3].set(edgecolor='black', linewidth=2)
texts[1].set(family='cursive', size=20)
autotexts[1].set(weight='bold', size=15)
autotexts[2].set(weight='bold', size=20)
autotexts[3].set(weight='bold', size=15)
```
