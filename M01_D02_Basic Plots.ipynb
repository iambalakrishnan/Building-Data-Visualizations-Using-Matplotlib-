```python
x = np.linspace(start=0, stop=10, num=50)
```

```python
plt.plot(x, np.sin(x))
plt.xlabel('x',fontsize=12, color='green')
plt.ylabel('sin(x)', fontsize=12, color='green')
```

```python
#we can also configure tick param values
plt.tick_params(axis='y',
               color='red',
               labelcolor='blue',
               labelsize='xx-large')
```

```python
#we can also use tick param for x axis
plt.tick_params(axis='x',
               color='red',
               labelcolor='green',
               labelsize='large')
```

```python
#No ticks on the x axis and no labels as well
plt.tick_params(axis='x',
               bottom=False,
               labelbottom=False)
```

```python
plt.plot(x, np.sin(x))
plt.xlabel('x',fontsize=12, color='green')
plt.ylabel('sin(x)', fontsize=12, color='green')
```

```python
#Legend will pickup the label which is defined in plot
plt.plot(x, np.sin(x), label='sine curve')
plt.legend()
plt.title('Sine plots')
```

If you have a large plot and if you want only a portion of the plot should be visible you can limit by passing xlim and ylim function

```python
plt.plot(x, np.sin(x), label='sine curve')
plt.legend()
plt.title('Sine plots')
plt.xlim(1,5)
```

```python
#xlim and ulim function 
plt.plot(x, np.sin(x), label='sine curve')
plt.legend()
plt.title('Sine plots')
plt.xlim(1,5)
plt.ylim(-1,0.5)
```

You can generate mirror image of your plot by switching the value of xlim parameters we are inverting the plot along with x axix

```python
#mirror image of your plot

plt.plot(x, np.sin(x), label='sine curve')
plt.legend()
plt.title('portion of Sine plot mirror')
plt.xlim(5,1)#switching values for mirror of plot along centre line of 3
plt.ylim(-1,0.5)
```

```python
# We can flip our figure along with horizontal as well using ylim parameters


plt.plot(x, np.sin(x), label='sine curve')
plt.legend()
plt.title('portion of Sine plot mirror')
plt.xlim(5,1)#switching values for mirror of plot along centre line of 3
plt.ylim(0.5,-1)
```
