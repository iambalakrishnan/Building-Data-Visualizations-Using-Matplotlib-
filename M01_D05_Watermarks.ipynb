```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
fig, ax = plt.subplots()

ax.plot([2,4,6,8],[4,8,12,16])

ax.text(4,10,'Line Plot',
       fontsize = 30,
        color = 'red',
       ha='left',
       va='bottom',
       alpha=0.5)
```

```python
x = np.linspace(start=0, stop=10, num=50)
```

```python
fig = plt.figure(figsize=(8,8)) 

ax = fig.add_subplot(221)

ax.plot([2,4,6,8],[4,8,12,16])

ax.set_label('straight line')

ax.text(4,10,'Line Plot',
       fontsize = 20,
        color = 'red',
       ha='left',
       va='bottom',
       alpha=0.5)
```

```python
fig = plt.figure(figsize=(8,8)) 

ax = fig.add_subplot(221)

ax.plot([2,4,6,8],[4,8,12,16])

ax.set_label('straight line')

ax.text(4,10,'Line Plot',
       fontsize = 20,
        color = 'red',
       ha='left',
       va='bottom',
       alpha=0.5)

ax2 = fig.add_subplot(222)
ax2.plot(x, np.sin(x))

ax3 = fig.add_subplot(223)
ax3.plot(x, np.cos(x))
```

```python
# A watermark for entire figure in addition to the one we have for one subplot
#By calling fig.text


fig = plt.figure(figsize=(8,8)) 

ax1 = fig.add_subplot(221)

ax1.plot([2,4,6,8],[4,8,12,16])

ax1.set_label('straight line')

ax1.text(4,10,'Line Plot',
       fontsize = 20,
        color = 'red',
       ha='left',
       va='bottom',
       alpha=0.5)

ax2 = fig.add_subplot(222)
ax2.plot(x, np.sin(x))

ax3 = fig.add_subplot(223)
ax3.plot(x, np.cos(x))

fig.text(0.2, 0.47,'Property of Plots',
       fontsize = 40,
        color = 'm',
       ha='left',
       va='bottom',
       alpha=0.5)
```
