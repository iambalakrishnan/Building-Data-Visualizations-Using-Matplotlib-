Twin axes can be used to represent same data in different units. Tos show temperature in fahrenheit and celsius, we define this conversion function 

```python
def fahrenheit2celsius(f):
    return (f - 32)*5/9
```

```python
fig, ax_tempF = plt.subplots()

fig.set_figwidth(12)
fig.set_figheight(6)

ax_tempF.set_xlabel('Date')
ax_tempF.tick_params(axis='x', bottom = False, labelbottom = False)

ax_tempF.set_ylabel('Temp(F)', color='red', size='x-large')
ax_tempF.tick_params(axis='y', labelcolor='red', labelsize='x-large')

ax_tempF.plot(austin_weather['Date'], austin_weather['TempAvgF'], color='green')

ax_tempC = ax_tempF.twinx()#Twin axes to display weather in Celsius

#We get the range of the first y axis- this will be used to set the range of the new Y axis 
ymin, ymax = ax_tempF.get_ylim()

#set the limits for the celsius axis by conveting the fahrenheit axis limits to celsius

ax_tempC.set_ylim((fahrenheit2celsius(ymin),fahrenheit2celsius(ymax)))

ax_tempC.set_ylabel('Temp(C)', color='blue', size='x-large')
ax_tempC.tick_params(axis = 'y', labelcolor='blue', labelsize='large')

```
