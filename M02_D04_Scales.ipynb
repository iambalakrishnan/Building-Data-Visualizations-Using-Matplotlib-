```python
import numpy as np
import pandas as pd
import matplotlib 
from matplotlib import pyplot as plt
```

```python
np.random.seed()

y = np.random.uniform(low = 0.0, high=1000, size=(1000,))
y.sort()
x = np.arange(len(y))
```

```python
plt.plot(x, y)
#Enabling grid function 
plt.grid(True)
#By default scale for a plot is linear, we just state it explicitly here 
plt.yscale('linear')


plt.show()
```

```python
plt.plot(x, y)
#Enabling grid function 
plt.grid(True)
#Adjust yscale to logarithmic 
plt.yscale('log')
#by default base for the log  scale is 10


plt.show()
```

```python
plt.plot(x, y)
#Enabling grid function 
plt.grid(True)
#By dault base for log is 10 and we set the log base to be 2
plt.yscale('log', basey=2)#log to the base 2

plt.show()
```

```python
plt.plot(x, y)
#Enabling grid function 
plt.grid(True)
#By dault base for log is 10 
plt.yscale('log', basey=10)

#setting up scale for x, set to base 2
plt.xscale('log', basex=2)

plt.show()
```
