```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3],[2, 4, 6])
ax.annotate('min value',#Text annotation which is rerutns to the screen
            xy = (1,2),#What data poin is being highlighted  
            xytext = (1, 3),#X and Y coordinates  of the bottom left of text annotations
            arrowprops = dict(facecolor = 'y', edgecolor='green', alpha=0.7))#Color of the arrow 



plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3],[2, 4, 6])
ax.annotate('Significant points',#Text annotation which is rerutns to the screen
            xy = (2,4),#What data poin is being highlighted  
            xytext = (2.5, 3),#X and Y coordinates  of the bottom left of text annotations
            arrowprops = dict(color = 'g'))#Color of the arrow 

ax.plot([2],[4],'ro')#Plot a red and circular mark on the line at (2,4)

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3],[2, 4, 6])
ax.annotate('Significant points',#Text annotation which is rerutns to the screen
            xy = (2,4),#What data poin is being highlighted  
            xytext = (2.5, 3),#X and Y coordinates  of the bottom left of text annotations
            arrowprops = dict(color = 'g',shrink=0.07))#shrink will fix the overlapping of the arrow over point 

ax.plot([2],[4],'go')#Plot a red and circular mark on the line at (2,4)

plt.show()
```

```python
np.random.seed(0)

x1 = -1 + np.random.randn(100)
y1 = -1 + np.random.randn(100)

x2 = 1 + np.random.randn(100)
y2 = 1 + np.random.randn(100)
```

```python
fig, ax = plt.subplots()

ax.scatter(x1, y1, color='r')
ax.scatter(x2, y2, color='b')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.scatter(x1, y1, color='r')
ax.scatter(x2, y2, color='b')

bbox_props = dict(boxstyle = 'square',
                 facecolor='w',
                 alpha=0.5)
ax.text(-2, -2,'Sample A', ha='center', va='center', size = 20, bbox = bbox_props)

ax.text(2, 2,'Sample B', ha='center', va='center', size = 20, bbox = bbox_props)


plt.show()
```

```python
fig, ax = plt.subplots()

ax.scatter(x1, y1, color='r')
ax.scatter(x2, y2, color='b')

bbox_props = dict(boxstyle = 'square',
                 facecolor='w',
                 alpha=0.5)
ax.text(-2, -2,'Sample A', ha='center', va='center', size = 15, bbox = bbox_props)

ax.text(2, 2,'Sample B', ha='center', va='center', size = 15, bbox = bbox_props)


arrow_bbox_probs = dict(boxstyle = 'rarrow',
                        facecolor = '#EBF5FB',
                        edgecolor = 'b',
                        linewidth = 2,
                        alpha = 0.7)
ax.text(0,0,'Direction', ha='center', va='center', rotation=50, size = 15, bbox = arrow_bbox_probs)

ax.text(-3.5, 2.75, 'Sample annotation', color='black', 
        bbox=dict(facecolor='none', edgecolor='black', boxstyle='round,pad=1'))

plt.show()
```
