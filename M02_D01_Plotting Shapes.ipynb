```python
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
from matplotlib import patches as patches
```

Drawing shapes with patches module, A patch in matplotlib represent 2D object

```python
fig, ax = plt.subplots()

#Rectangle patch is built in patch in matplotlib patches 
ax.add_patch(patches.Rectangle((0.1, 0.1), 0.5, 0.5, fill = False))

plt.show()
```

```python
#Setting aspect of both x and y axis 

fig, ax = plt.subplots()
ax.set_aspect(aspect = 'equal')
#Rectangle patch is built in patch in matplotlib patches 
ax.add_patch(patches.Rectangle((0.1, 0.1), 0.5, 0.5, fill = False))

plt.show()

#This look more like square 
```

```python
 #Setting aspect of both x and y axis 

fig, ax = plt.subplots()
#ax.set_aspect(aspect = 'equal')
#Rectangle patch is built in patch in matplotlib patches 
#you can also customize patch color
ax.add_patch(patches.Rectangle((0.1, 0.1), 0.5, 0.5, facecolor = 'yellow', edgecolor = 'green'))

plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect = 'equal')
for p in [
    patches.Rectangle((0.1,0.1),0.3, 0.6, hatch='.'),
    patches.Rectangle((0.5, 0.1), 0.3, 0.6, hatch='\\', fill = False),
]:
    ax.add_patch(p)

#The hatch attribute sets the pattern with which to fill a patch
plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect = 'equal')
for p in [
    patches.Rectangle((0.1,0.1),0.2, 0.6, alpha=None,),
    patches.Rectangle((0.4, 0.1), 0.2, 0.6, alpha=1.0),#alpha = None and alpha=1 both are same
    patches.Rectangle((0.7,0.1),0.2, 0.6, alpha=0.6,),
    patches.Rectangle((1.0, 0.1), 0.2, 0.6, alpha=0.3),
]:
    ax.add_patch(p)
    ax.set_xlim(0, 1.5)

#The hatch attribute sets the pattern with which to fill a patch
plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect = 'equal')
for p in [
    #There are different ways to specify fill color 
    #By default matplotlib picks color in a sequence from the default color palette, the first color is blue
    patches.Rectangle((0.1,0.1),0.2, 0.6, facecolor=None,),
    #The string none-'none' means there is no color at all
    patches.Rectangle((0.4, 0.1), 0.2, 0.6, facecolor='none'),
    patches.Rectangle((0.7,0.1),0.2, 0.6, facecolor='blue',),
    patches.Rectangle((1.0, 0.1), 0.2, 0.6, facecolor='#00ffff'),
]:
    ax.add_patch(p)
    ax.set_xlim(0, 1.5)

#The hatch attribute sets the pattern with which to fill a patch
plt.show()
```

```python
#We can aslo leave the shapes hollow and fill the edges

fig, ax = plt.subplots()
ax.set_aspect(aspect = 'equal')
for p in [
    #There are different ways to specify fill color 
    #By default matplotlib picks color in a sequence from the default color palette, the first color is blue
    patches.Rectangle((0.1,0.1),0.2, 0.6, fill=False, edgecolor=None,),
    #The string none-'none' means there is no color at all
    patches.Rectangle((0.4, 0.1), 0.2, 0.6, fill=False, edgecolor='none'),
    patches.Rectangle((0.7,0.1),0.2, 0.6, fill=False, edgecolor='blue',),
    patches.Rectangle((1.0, 0.1), 0.2, 0.6, fill=False, edgecolor='#00ffff'),
]:
    ax.add_patch(p)
    ax.set_xlim(0, 1.5)

#The hatch attribute sets the pattern with which to fill a patch
plt.show()
```

```python
#the edges can be treated as lines and formatted accrodingly 

fig, ax = plt.subplots()
ax.set_aspect(aspect = 'equal')
for p in [
    #There are different ways to specify fill color 
    #By default matplotlib picks color in a sequence from the default color palette, the first color is blue
    patches.Rectangle((0.1,0.1),0.2, 0.6, fill=False, linestyle='solid'),#Default line style is solid
    #The string none-'none' means there is no color at all
    patches.Rectangle((0.4, 0.1), 0.2, 0.6, fill=False, linestyle='dashed'),
    patches.Rectangle((0.7,0.1),0.2, 0.6, fill=False, linestyle='dashdot'),
    patches.Rectangle((1.0, 0.1), 0.2, 0.6, fill=False, linestyle='dotted'),
]:
    ax.add_patch(p)
    ax.set_xlim(0, 1.5)

#The hatch attribute
```

```python
#We can draw any type of patch object 


fig, ax = plt.subplots()
ax.set_aspect(aspect = 'equal')
for p in [
    patches.Circle((0.1,0.4),0.1, hatch='/'),
    patches.Circle((0.5,0.4),0.1, hatch='*', facecolor='red'),
    patches.Circle((0.9,0.4),0.1, hatch='\\', facecolor='green'),
    patches.Circle((0.5,0.7),0.1, hatch='/', fill = False),
]:
    ax.add_patch(p)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.set_aspect(aspect='equal')
#You need to specify the x and y coordinates of each vertices of the polygon 
polygon = patches.Polygon([[0.1, 0.1],
                            [0.2, 0.8],
                            [0.5, 0.7],
                            [0.7, 0.1],
                           # [0.9,0.2],
                            [0.4, 0.3]], 
                           fill=False)
ax.add_patch(polygon)
plt.show()
```

```python
#By default matplotlib assumes you want to close th polygon and it connect the last vertex to the first

#If you want explicitly leave your polygon open you can specify it by calling closed=False method

fig, ax = plt.subplots()

ax.set_aspect(aspect='equal')
#You need to specify the x and y coordinates of each vertices of the polygon 
polygon = patches.Polygon([[0.1, 0.1],
                            [0.2, 0.8],
                            [0.5, 0.7],
                            [0.7, 0.1],
                            [0.4, 0.3]],
                            closed=False,
                           fill=False)
ax.add_patch(polygon)
plt.show()
```

```python
#Lets draw arrow

fig, ax = plt.subplots()

ax.set_aspect(aspect='equal')
#You need to specify the x and y coordinates of each vertices of the polygon 
polygon = patches.Arrow(0.1, 0.2, 0.5, 0.5)
ax.add_patch(polygon)
plt.show()
```
