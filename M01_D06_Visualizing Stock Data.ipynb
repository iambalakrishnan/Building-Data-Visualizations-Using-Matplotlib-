```python
import numpy as np
import pandas as pd
import matplotlib 
from matplotlib import pyplot as plt
```

```python
#Stocks.csv contains closing price of each stocks on the first trading day of each month from Jan 2007 to Jan 2017
stock_data = pd.read_csv('stocks.csv', parse_dates=['Date'])
stock_data.head(3)
```

```python
stock_data.info()
```

```python
fig = plt.figure(figsize=(10,6))

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.05, 0.65, 0.5, 0.3])
```

```python
#Lets plot apple's closing price 
fig = plt.figure(figsize=(10,6))

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.05, 0.65, 0.5, 0.3])

ax1.plot(stock_data['Date'], stock_data['AAPL'], color='green')
ax1.set_title('AAPL vs IBM(inset)')

ax2.plot(stock_data['Date'], stock_data['IBM'], color='blue')
```

```python
#Lets plot multiple stock performance side by side

fig = plt.figure(figsize=(10, 6))

#Title for entire figure
fig.suptitle('Stock price comparision 2007 - 2017', fontsize = 20)

ax1 = fig.add_subplot(221)
ax1.set_title('Microsoft')
ax1.plot(stock_data['Date'], stock_data['MSFT'], color='green')

ax2 = fig.add_subplot(222)
ax2.set_title('Google')
ax2.plot(stock_data['Date'],stock_data['GOOG'], color='purple')

ax3=fig.add_subplot(223)
ax3.set_title('Apple')
ax3.plot(stock_data['Date'], stock_data['AAPL'], color='blue')

ax4 = fig.add_subplot(224)
ax4.set_title('IBM')
ax4.plot(stock_data['Date'], stock_data['IBM'], color='red')

fig.tight_layout()
```
